```python
# Detect missing values (NaN in numeric arrays, None/NaN in object arrays) for "RIDEXMON"
demo.loc[demo.RIDEXMON.isnull()]
```

```python
# remove the rows which "RIDEXMON" and "RIDEXAGM" is NaN
demo.loc[((~demo.RIDEXMON.isnull()) & (~demo.RIDEXAGM.isnull()))]
```

```python
merged = bmx.merge(right = demo, on = 'SEQN')
merged.head()
```

```python
merged2 = demo.merge(right = bmx, on = 'SEQN')
merged2.head()
```

```python
merged.info()
```

```python
## rows that do not have missing values for the variable AGE in months. persons between 0 and 24 months
babies = merged.loc[merged['RIDAGEMN'].isnull() == False]
```

```python
# or
## rows that do not have missing values for the variable AGE in months. persons between 0 and 24 months
babies = merged.loc[~merged.RIDAGEMN.isnull()]
```

```python
print(babies.loc[:, ['RIDAGEMN', 'BMXWT']].info())
# only have 661 babies with body weight measurements
```

```python
# or 
print(babies.loc[:, ('RIDAGEMN', 'BMXWT')].info())
# only have 661 babies with body weight measurements
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(babies.RIDAGEMN, babies.BMXWT)
plt.show()
```

```python
babies.loc[ np.round(babies['RIDAGEMN'], 2) == 0, "RIDAGEMN"].head()
```

```python
# simple example for np.round
print(np.round(1.23456789,2))
print(np.round(5.397605e-79,2)==0)
```

```python
# average weight of a 0 month old
# we had to use np.round because testing for direct equality to 0 resulted in 0 values returned
babies.loc[ np.round(babies['RIDAGEMN'], 2) == 0 , 'BMXWT'].plot(kind = 'box')
```

```python
print(babies['RIDAGEMN'].astype(int)==0)
```

```python
# We could have also converted Age in months to integers
babies.loc[ babies['RIDAGEMN'].astype(int) == 0 , 'BMXWT'].plot(kind = 'box')
plt.ylabel("mean")
plt.show()
print(babies.loc[ babies['RIDAGEMN'].astype(int) == 0 , 'BMXWT'].mean()) # mean weight of babies that are 0 months old
```

```python
sns.lmplot(x = 'RIDAGEMN', y = 'BMXWT', hue = 'RIDRETH3', data = babies)
```

```python
sns.lmplot(x = 'RIDAGEMN', y = 'BMXWT', hue = 'RIDRETH3', col = 'RIDRETH3', data = babies)  # col stand for columns 
```
